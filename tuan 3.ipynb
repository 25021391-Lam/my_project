#Viết một chương trình nhập vào một số nguyên n và in ra các chữ số đảo ngược nhau

```python
n = int(input())
so_dao_nguoc = 0
x = abs(n)
while x > 0:
    chu_so_cuoi = x % 10
    so_dao_nguoc = (so_dao_nguoc * 10) + chu_so_cuoi
    x = x //10
if n <0:
    so_dao_nguoc = -so_dao_nguoc
print(so_dao_nguoc)

```

Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0 


```python

def luy_thua_cua_2(n):
    if n < 0:
        print('False')
    elif n>0 and n&(n-1)==0:
        print('True')
    else:
        print('False')
so = int(input())
luy_thua_cua_2(so)        
```

[Tính tiền điện]
Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh 


```python
n = int(input())
if 0<= n <=50:
    tien_dien= n*1500
    print(tien_dien)
elif 51<=n<=100:
    tien_dien=50*1500+ (n-50)*2000
    print(tien_dien)
elif n>100:
    tien_dien=50*1500+50*2000+(n-100)*3000
    print(tien_dien)
```

[Rounding]
Nhập vào một số thực: in ra kết quả làm tròn lên ,tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm
sẵn có của pyton). Các kết quả cách nhau một dấu cách 

```python
x = float(input())

tron_xuong = int(x // 1)
tron_len = int(-(-x // 1))
tron_gan_nhat = int((x * 2 + 1) // 2) 

print(f"{tron_len} {tron_xuong} {tron_gan_nhat}")
```

Cấp số nhân
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng
thứ tự a, b, c, d hay không?
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội và so sánh vs c, c nhân công bội và so sánh vs
d. 


```python
a, b, c, d = map(int, input().split())
if a!=0 and b%a ==0:
    q=b//a
    if b*q==c and c*q==d:
        print('Yes')
    else:
        print('No')
elif a==0 and b==0 and c==0 and d==0:
    print('Yes')
else:
    print('No')
```
